```python
def gerar_vendas(min_val=100, max_val=700, quantidade=4):
    return [random.randint(min_val, max_val) for _ in range(quantidade)]
```

```python
dados_vendas = {'Produto': ['A', 'B', 'C', 'D'],
                  'Vendas Jan': gerar_vendas(),
                  'Vendas Fev': gerar_vendas(),
                  'Vendas Mar': gerar_vendas()[:-1] + [random.randint(155, 655)],
                  'Vendas Abr': gerar_vendas(),
                  'Vendas Mai': gerar_vendas(),
                  'Vendas Jun': gerar_vendas(),
                  'Vendas Jul': gerar_vendas(155, 655),
                  'Vendas Ago': gerar_vendas(222, 488),
                  'Vendas Set': gerar_vendas(333, 666),
                  'Vendas Out': gerar_vendas(),
                  'Vendas Nov': gerar_vendas(),
                  'Vendas Dez': gerar_vendas(155, 655)}







df_vendas = pd.DataFrame(dados_vendas)
sns.set_style('whitegrid')

sns.barplot(x='Produto', y='Vendas Jan', data=df_vendas, color='blue')


vendas_scatter = pd.DataFrame({'Vendas': gerar_vendas(),
                               'Gastos': gerar_vendas(),
                               'Média de preço': gerar_vendas()
                               })
```

```python
df_vendas = pd.DataFrame(dados_vendas)
sns.set_style('whitegrid')

sns.barplot(x='Produto', y='Vendas Jan', data=df_vendas, color='blue')


vendas_scatter = pd.DataFrame({'Vendas': gerar_vendas(),
                               'Gastos': gerar_vendas(),
                               'Média de preço': gerar_vendas()
                               })


plt.scatter(x="Gastos", y="Vendas", c='Média de preço', cmap='blue', data=vendas_scatter)
plt.title('Vendas vs Gastos')
plt.xlabel('Vendas(Em unidades)')
plt.colorbar().set_label('Média de preço')
plt.show()
```
