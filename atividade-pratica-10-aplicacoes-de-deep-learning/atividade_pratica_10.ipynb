```python
# Instalando o pacote de compatibilidade necessário
!pip install tf_keras --quiet

import os
os.environ["TF_USE_LEGACY_KERAS"] = "1"

import tf_keras as keras
from PIL import Image, ImageOps
import numpy as np

# Desativar notação científica
np.set_printoptions(suppress=True)

# Carregar o modelo usando a versão de compatibilidade
model = keras.models.load_model("keras_model.h5", compile=False)

# Carregar as legendas
class_names = open("labels.txt", "r").readlines()

# Preparar o array para o modelo
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

# Caminho da imagem
image_path = "/content/animais/teste1.png"
image = Image.open(image_path).convert("RGB")

# Redimensionar e cortar a imagem
size = (224, 224)
image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)

# Converter para array e normalizar
image_array = np.asarray(image)
normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1

# Carregar no array de entrada
data[0] = normalized_image_array

# Predição
prediction = model.predict(data)
index = np.argmax(prediction)
class_name = class_names[index]
confidence_score = prediction[0][index]

# Resultado
print("\nClasse:", class_name[2:].strip())
print("Score de Confiança:", f"{confidence_score:.2%}")
```
