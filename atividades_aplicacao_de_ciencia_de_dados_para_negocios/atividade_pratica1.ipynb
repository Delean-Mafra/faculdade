```python
# 1. Importações
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações visuais
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# 1.b Dados fictícios e DataFrame
data = {
    'Data': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'Produto': ['Produto A', 'Produto B', 'Produto C'],
    'Quantidade Vendida': [30, 20, 15],
    'Receita': [3000, 2000, 1500],
    'Custo': [2000, 1200, 800]
}

df = pd.DataFrame(data)

# 1.c Coluna Lucro
df['Lucro'] = df['Receita'] - df['Custo']

# Converter Data para datetime (útil para gráficos de série temporal)
df['Data'] = pd.to_datetime(df['Data'])

# Exibir DataFrame
df
```

```python
# 2.a Totais de receita, custo e lucro
total_receita = df['Receita'].sum()
total_custo = df['Custo'].sum()
total_lucro = df['Lucro'].sum()

print(f"Total Receita: {total_receita}")
print(f"Total Custo: {total_custo}")
print(f"Total Lucro: {total_lucro}")

# 2.b Margem de Lucro (%)
df['Margem Lucro (%)'] = (df['Lucro'] / df['Receita']) * 100
df['Margem Lucro (%)'] = df['Margem Lucro (%)'].round(2)

# Mostrar DataFrame com a nova coluna
df
```

```python
# 2.c Filtrar produtos com margem de lucro acima de 30%
produtos_lucrativos = df[df['Margem Lucro (%)'] > 30]
produtos_lucrativos
```

```python
plt.figure()
sns.barplot(x='Produto', y='Quantidade Vendida', data=df, palette='pastel')
plt.title('Quantidade Vendida por Produto')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
plt.figure()
sns.lineplot(x='Data', y='Lucro', data=df, marker='o')
plt.title('Evolução do Lucro')
plt.xlabel('Data')
plt.ylabel('Lucro (R$)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
plt.pie(df['Receita'], labels=df['Produto'], autopct='%1.1f%%', colors=['#66c2a5','#fc8d62','#8da0cb'])
plt.title('Participação na Receita por Produto')
plt.show()
```

```python
# Código completo reunido
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (9, 6)

data = {
    'Data': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'Produto': ['Produto A', 'Produto B', 'Produto C'],
    'Quantidade Vendida': [30, 20, 15],
    'Receita': [3000, 2000, 1500],
    'Custo': [2000, 1200, 800]
}

df = pd.DataFrame(data)
df['Data'] = pd.to_datetime(df['Data'])
df['Lucro'] = df['Receita'] - df['Custo']
df['Margem Lucro (%)'] = ((df['Lucro'] / df['Receita']) * 100).round(2)

# Totais
total_receita = df['Receita'].sum()
total_custo = df['Custo'].sum()
total_lucro = df['Lucro'].sum()

print(f"Total Receita: {total_receita}")
print(f"Total Custo: {total_custo}")
print(f"Total Lucro: {total_lucro}")
print("\nDataFrame:\n", df)

# Filtrar produtos com margem > 30%
produtos_lucrativos = df[df['Margem Lucro (%)'] > 30]
print("\nProdutos com margem > 30%:\n", produtos_lucrativos[['Produto','Margem Lucro (%)']])

# Visualizações
plt.figure()
sns.barplot(x='Produto', y='Quantidade Vendida', data=df, palette='pastel')
plt.title('Quantidade Vendida por Produto')
plt.show()

plt.figure()
sns.lineplot(x='Data', y='Lucro', data=df, marker='o')
plt.title('Evolução do Lucro')
plt.xticks(rotation=45)
plt.show()

plt.figure()
plt.pie(df['Receita'], labels=df['Produto'], autopct='%1.1f%%', colors=['#66c2a5','#fc8d62','#8da0cb'])
plt.title('Participação na Receita por Produto')
plt.show()
```
