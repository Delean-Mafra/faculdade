```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar o arquivo Excel
# Ajuste o caminho conforme necessário se estiver no Colab ou Jupyter
df = pd.read_excel("Pasta1.xlsx")

# 2. Visualizar os primeiros registros
print(df.head())

# 3. Gráfico de colunas - Vendas por Cidade
plt.figure(figsize=(10,6))
sns.barplot(x="Cidade", y="Vendas", data=df, estimator=sum, ci=None)
plt.title("Vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total de Vendas")
plt.show()

# 4. Gráfico de pizza - Distribuição da Satisfação
satisfacao_counts = df["Satisfação"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(satisfacao_counts, labels=satisfacao_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribuição da Satisfação dos Clientes")
plt.show()

# 5. Gráfico de linha - Tendência de Vendas ao longo do tempo
plt.figure(figsize=(10,6))
sns.lineplot(x="Data", y="Vendas", data=df, marker="o")
plt.title("Tendência de Vendas ao Longo do Tempo")
plt.xlabel("Data")
plt.ylabel("Vendas")
plt.show()

# 6. Histograma - Distribuição das Vendas
plt.figure(figsize=(8,6))
sns.histplot(df["Vendas"], bins=5, kde=True)
plt.title("Distribuição das Vendas")
plt.xlabel("Vendas")
plt.ylabel("Frequência")
plt.show()

# 7. Gráfico de dispersão - Produtos x Vendas (se houver coluna Produtos)
if "Produtos" in df.columns:
    plt.figure(figsize=(8,6))
    sns.scatterplot(x="Produtos", y="Vendas", data=df)
    plt.title("Relação entre Produtos Vendidos e Vendas")
    plt.xlabel("Produtos Vendidos")
    plt.ylabel("Vendas")
    plt.show()
```

```python
data = pd.date_range(start='2023-01-01', periods=12, freq='M')
vendas = [1500, 1800, 1700, 1900, 2100, 2200, 2000, 2300, 2500, 2400, 2600, 2700]

df = pd.DataFrame({'Data': data, 'Vendas': vendas})
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='Data', y='Vendas', data=df, marker='o')
plt.title('Tendência de Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Vendas'], bins=5, kde=True)
plt.title('Distribuição das Vendas Mensais')
plt.xlabel('Vendas')
plt.ylabel('Frequência')
plt.show()
```

```python
produtos = [100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 320]
df['Produtos'] = produtos

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Produtos', y='Vendas', data=df)
plt.title('Relação entre Produtos Vendidos e Vendas')
plt.xlabel('Produtos Vendidos')
plt.ylabel('Vendas')
plt.show()
```
