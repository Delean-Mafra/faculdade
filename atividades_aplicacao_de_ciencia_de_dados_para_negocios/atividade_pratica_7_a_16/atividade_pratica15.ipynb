# Atividade Prática 15 – Análise de Textos com Tokenização e Count Vectorization

Este notebook implementa o processo de tokenização e vetorização de textos utilizando as bibliotecas `nltk` e `scikit-learn`.

```python
# Instalação das bibliotecas necessárias
!pip install nltk scikit-learn
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from sklearn.feature_extraction.text import CountVectorizer

# Download dos recursos necessários para a tokenização
nltk.download('punkt')
nltk.download('punkt_tab')
```

## Passo 1: Tokenização de Texto
Vamos dividir o texto em unidades menores (tokens) como palavras e sentenças.

```python
texto = "Tokenização é um passo fundamental em processamento de linguagem natural. Ela é essencial para análise de sentimentos."

# Agora com os recursos baixados, a tokenização funcionará
palavras_tokenizadas = word_tokenize(texto)
sentencas_tokenizadas = sent_tokenize(texto)

print("Texto original:", texto)
print("\nPalavras tokenizadas:", palavras_tokenizadas)
print("\nSentenças tokenizadas:", sentencas_tokenizadas)
```

## Passo 2: Count Vectorization
Agora, vamos transformar os documentos em vetores numéricos baseados na frequência das palavras.

```python
documentos = [
    "Tokenização é um passo fundamental em processamento de linguagem natural.",
    "Ela é essencial para análise de sentimentos e outros processos de dados."
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documentos)

print("Palavras do vocabulário:", vectorizer.get_feature_names_out())
print("\nRepresentação vetorial dos documentos:\n", X.toarray())
```

## Análise dos Resultados

**1. Quais são as palavras tokenizadas no texto?**
As palavras extraídas no primeiro exemplo foram: 'Tokenização', 'é', 'um', 'passo', 'fundamental', 'em', 'processamento', 'de', 'linguagem', 'natural', '.', 'Ela', 'é', 'essencial', 'para', 'análise', 'de', 'sentimentos', '.'.

**2. Como a Count Vectorization representa os documentos numericamente?**
A Count Vectorization cria uma matriz onde cada linha representa um documento e cada coluna representa uma palavra única do vocabulário total. O valor em cada célula é o número de vezes que aquela palavra aparece no documento.

**3. O que você pode inferir sobre a frequência das palavras nos documentos?**
Podemos observar quais palavras são mais comuns em cada frase. No exemplo, a palavra 'de' aparece duas vezes no segundo documento, o que é refletido no vetor correspondente.

## 🔎 Melhorias na Análise de Feedbacks

Para obter insights mais precisos, vamos:
1. Remover **stopwords** (palavras que não agregam valor semântico como 'o', 'a', 'foi').
2. Criar uma **visualização gráfica** das frequências.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Updated stopwords list to exclude irrelevant terms
stopwords_pt = ['o', 'a', 'e', 'do', 'da', 'foi', 'um', 'uma', 'para', 'com', 'no', 'na', 'mas', 'estou', 'não', 'de', 'em']

vectorizer_refined = CountVectorizer(stop_words=stopwords_pt)
X_refined = vectorizer_refined.fit_transform(feedbacks)

# Creating a DataFrame for visualization
vocab_refined = vectorizer_refined.get_feature_names_out()
counts_refined = X_refined.toarray().sum(axis=0)

df_frequencia = pd.DataFrame({'Palavra': vocab_refined, 'Frequência': counts_refined})
df_frequencia = df_frequencia.sort_values(by='Frequência', ascending=False)

# Plotting results - Fix for FutureWarning
plt.figure(figsize=(10, 6))
sns.barplot(x='Frequência', y='Palavra', data=df_frequencia, hue='Palavra', palette='viridis', legend=False)
plt.title('Palavras mais frequentes nos Feedbacks (Sem Stopwords)')
plt.xlabel('Ocorrências')
plt.ylabel('Termos')
plt.show()
```
