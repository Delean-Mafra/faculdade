```python
import pandas as pd

# Dados de exemplo
dados = {
    'Meses': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    'Eletrônicos': [1200, 1300, 1250, 1400, 1500, 1600, 1700, 1650, 1800, 1900, 2000, 2100],
    'Roupas': [900, 950, 920, 1000, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1450],
    'Alimentos': [600, 620, 610, 630, 650, 670, 680, 690, 700, 710, 720, 730],
}

df = pd.DataFrame(dados)
display(df.head())
```

## 1. Gráfico de Linha: Tendência de Vendas Mensais por Categoria

Este gráfico de linha mostra a evolução das vendas de cada categoria de produto ao longo dos meses do ano.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
for categoria in ['Eletrônicos', 'Roupas', 'Alimentos']:
    sns.lineplot(x='Meses', y=categoria, data=df, marker='o', label=categoria)

plt.title('Tendência de Vendas Mensais por Categoria')
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.legend(title='Categorias')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

### Análise do Gráfico de Linha:

*   **Tendências de Vendas**: Todas as categorias mostram uma tendência geral de crescimento nas vendas ao longo do ano, com um aumento mais notável no segundo semestre, possivelmente indicando sazonalidade.
*   **Performance por Categoria**: 'Eletrônicos' consistentemente apresenta os maiores volumes de vendas, seguido por 'Roupas' e, por último, 'Alimentos'.
*   **Sazonalidade**: Observa-se um crescimento mais acentuado para 'Eletrônicos' e 'Roupas' nos últimos meses do ano (Out, Nov, Dez), o que pode estar relacionado a datas comemorativas ou promoções de fim de ano.

## 2. Gráfico de Barras: Vendas por Categoria em um Mês Específico (Janeiro)

Este gráfico compara o desempenho de vendas de cada categoria para o mês de Janeiro.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
# Seleciona os dados do primeiro mês (Janeiro) para as categorias especificadas
df_jan = df[['Eletrônicos', 'Roupas', 'Alimentos']].iloc[0]
df_jan.plot(kind='bar', color=['skyblue', 'lightgreen', 'lightcoral'])

plt.title('Vendas por Categoria em Janeiro')
plt.xlabel('Categorias')
plt.ylabel('Vendas')
plt.xticks(rotation=0) # Ensure labels are horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Análise do Gráfico de Barras:

*   **Comparação de Performance**: Em Janeiro, 'Eletrônicos' lidera as vendas, seguido por 'Roupas' e 'Alimentos'. Essa proporção de vendas entre as categorias parece ser consistente com o observado no gráfico de linha.
*   **Insights**: Os dados de Janeiro servem como um bom ponto de partida para analisar o desempenho inicial do ano, confirmando a dominância de 'Eletrônicos' e a menor contribuição de 'Alimentos' em termos de volume de vendas.

## 3. Gráfico de Pizza: Participação de Mercado por Categoria no Total de Vendas Anuais

Este gráfico mostra a proporção do total de vendas anuais que cada categoria representa, indicando sua participação de mercado.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 9))
# Calcula a soma total de vendas para cada categoria ao longo do ano
pie_data = df[['Eletrônicos', 'Roupas', 'Alimentos']].sum()
pie_labels = ['Eletrônicos', 'Roupas', 'Alimentos']
pie_colors = ['gold', 'yellowgreen', 'lightcoral']

plt.pie(pie_data,
        labels=pie_labels,
        autopct='%1.1f%%',
        colors=pie_colors,
        startangle=90,
        pctdistance=0.85) # Adjust percentage distance for better readability

plt.title('Participação de Mercado por Categoria (Vendas Anuais Totais)')
plt.axis('equal') # Ensures the pie chart is circular.
plt.tight_layout()
plt.show()
```

### Análise do Gráfico de Pizza:

*   **Distribuição de Vendas**: O gráfico de pizza reforça que 'Eletrônicos' detém a maior fatia do mercado em termos de vendas anuais totais, seguido por 'Roupas' e 'Alimentos'.
*   **Identificação de Dominância**: 'Eletrônicos' é claramente a categoria mais dominante. Esta informação é crucial para decisões estratégicas, como alocação de recursos, campanhas de marketing e planejamento de estoque, onde o foco pode ser mantido nas categorias de maior contribuição ou em estratégias para impulsionar as categorias de menor participação.
